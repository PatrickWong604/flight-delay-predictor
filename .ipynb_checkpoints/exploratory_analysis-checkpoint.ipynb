#### **Task 8**: When (which hour) do most 'LONG', 'SHORT', 'MEDIUM' haul flights take off?

```python
df_flights = df_flights[df_flights['cancelled']!=1]
```

```python
df_flights = df_flights[df_flights['diverted']!=1]
```

```python
df_flights
```

```python
df_flights['air_time_hr'] = df_flights['air_time']/60
```

```python
df_flights['air_time_hr']
```

```python
# air time: short < 3, medium 3-7, long > 7
def air_time(df):
    '''
    Bin air time into lengths
    Short: 3 <
    Medium: 3 - 7
    Long: > 7
    '''
    label_names = ['Short','Medium','Long']
    cut_points = [0,3,7,np.inf]
    df_flights["air_time_bin"] = pd.cut(df_flights["air_time_hr"], cut_points, labels=label_names,include_lowest=True)
    return df
```

```python
air_time(df_flights)
```

```python
fig, ax = plt.subplots(figsize=(12,8))
sns.barplot(y=df_flights['air_time_hr'],x=df_flights['dep_time_hr'],hue=df_flights['air_time_bin'])
plt.xticks(rotation=45)
plt.ylabel('Air Time')
plt.xlabel('Departure Time')
plt.title('Air Time by Departure Time (hour)')
plt.show()
```

#### **Task 10**: Do bigger delays lead to bigger fuel comsumption per passenger? 
We need to do four things to answer this as accurate as possible:
- Find out average monthly delay per air carrier (monthly delay is sum of all delays in 1 month)
- Find out distance covered monthly by different air carriers
- Find out number of passengers that were carried by different air carriers
- Find out total fuel comsumption per air carrier.

Use this information to get the average fuel comsumption per passenger per km. Is this higher for the airlines with bigger average delays?

```python
#Add month and year
df_flights['month'] = pd.to_datetime(df_flights['fl_date']).dt.month
df_flights['year'] = pd.to_datetime(df_flights['fl_date']).dt.year
```

Find out average monthly delay per air carrier (monthly delay is sum of all delays in 1 month)

```python
#average monthly dep_delay per air carrier
avg_delay_per_carrier = df_flights[['op_unique_carrier', 'month', 'dep_delay']].groupby(['op_unique_carrier', 'month']).mean().reset_index()
avg_delay_per_carrier
```

Find out distance covered monthly by different air carriers

```python
#distance covered monthly by different air carriers
dist_monthly_per_carrier = df_flights[['op_unique_carrier', 'year', 'month', 'distance']].groupby(['op_unique_carrier', 'year', 'month']).sum().reset_index()
dist_monthly_per_carrier
```

Find out number of passengers that were carried by different air carriers

```python
passengers = pd.read_csv('passenger500000.csv')
```

```python
#number of passengers that were carried by different air carriers
num_pass_per_carrier = passengers[['unique_carrier', 'passengers']].groupby(['unique_carrier']).sum().reset_index()
num_pass_per_carrier
```

Find out total fuel comsumption per air carrier.

```python
fuel = pd.read_csv('fuel.csv')
fuel.head()
```

```python
#filter out by year 2019
fuel = fuel[fuel['year']==2019]
```

```python
#total fuel comsumption per air carrier
fuel_per_carrier = fuel[['unique_carrier', 'total_gallons']].groupby(['unique_carrier']).sum().reset_index()
fuel_per_carrier.head()
```

Use this information to get the average fuel comsumption per passenger per km. Is this higher for the airlines with bigger average delays?

```python
#Merge fuel and passenger and drop nas
fuel_per_passenger = pd.merge(fuel_per_carrier, num_pass_per_carrier, how='left', on='unique_carrier')
#Drop nas
fuel_per_passenger = fuel_per_passenger.dropna()
#Drop zeros
fuel_per_passenger = fuel_per_passenger[(fuel_per_passenger != 0).all(1)]
fuel_per_passenger.head()
```

```python
#Calc total distance per carrier from flights table
dist_per_carrier = df_flights[['op_unique_carrier', 'distance']].groupby(['op_unique_carrier']).sum().reset_index()
dist_per_carrier.head()
```

```python
# Merge fuel_per_passenger and distance
fuel_per_passenger_per_dist = pd.merge(fuel_per_passenger, dist_per_carrier, how='left', left_on='unique_carrier', right_on='op_unique_carrier')
fuel_per_passenger_per_dist = fuel_per_passenger_per_dist.dropna()
fuel_per_passenger_per_dist = fuel_per_passenger_per_dist.drop(['op_unique_carrier'], axis=1)
fuel_per_passenger_per_dist
```

```python
# Calc average fuel comsumption per passenger per km
# Convert miles to km
fuel_per_passenger_per_dist['avg_fuel_per_pass_per_km'] = fuel_per_passenger_per_dist['total_gallons'] / fuel_per_passenger_per_dist['passengers'] / (fuel_per_passenger_per_dist['distance'] * 1.60934)
fuel_per_passenger_per_dist = fuel_per_passenger_per_dist.reset_index(drop=True)
fuel_per_passenger_per_dist
```

```python
#Avg delay per carrier
avg_delay_per_carrier = df_flights[['op_unique_carrier','dep_delay']].groupby(['op_unique_carrier']).mean().reset_index()
avg_delay_per_carrier
```

```python
#merge to fuel_per_passengers_dist table
fuel_per_passenger_per_dist  = pd.merge(fuel_per_passenger_per_dist,avg_delay_per_carrier, how='left', left_on='unique_carrier',right_on='op_unique_carrier')
fuel_per_passenger_per_dist
```

```python
fig, ax = plt.subplots(figsize=(12,8))
sns.scatterplot(data=fuel_per_passenger_per_dist, x='dep_delay',y='avg_fuel_per_pass_per_km',hue='unique_carrier')
plt.legend(loc='upper right')
plt.show()
```

```python
sns.lineplot(data=fuel_per_passenger_per_dist, x='dep_delay',y='avg_fuel_per_pass_per_km')
```
