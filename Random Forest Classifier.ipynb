```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns',False)
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score
from sklearn.metrics import recall_score, precision_score, r2_score, accuracy_score, f1_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import StandardScaler
```

```python
df_flights = pd.read_csv('cleaned_flights_sample2.csv')
```

```python
df_flights['arr_delay2'] = np.where(df_flights['arr_delay'] > 0, 1,
                    np.where(df_flights['arr_delay'] <= 0, 0,0))
```

```python
df_flights = df_flights.drop(['fl_date','crs_dep_bin','crs_arr_bin','arr_delay','departures_performed','payload','seats','passengers','freight','mail','ramp_to_ramp'],axis=1)
```

```python
df_flights.head()
```

```python
df_flights = pd.get_dummies(df_flights)
```

```python
X = df_flights.drop(['arr_delay2'],axis=1)
y = df_flights['arr_delay2']
```

```python
scaler = StandardScaler()
X_scale = scaler.fit_transform(X)
X_scale = pd.DataFrame(X_scale,columns=X.columns)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scale,y,test_size=0.3)
```

```python
rf = RandomForestClassifier(n_estimators=500, max_depth=10)
```

```python
rf.fit(X_train,y_train)
```

```python
rf_pred = rf.predict(X_test)
```

```python
print("Accuracy:",accuracy_score(y_test, rf_pred))
print("Precision:",precision_score(y_test, rf_pred,average='macro'))
print("Recall:",recall_score(y_test, rf_pred,average='macro'))
print("F1:",f1_score(y_test, rf_pred,average='macro'))
```

```python
confusion_matrix(y_test,rf_pred)
```
